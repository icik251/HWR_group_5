Running SIFT and Harris Corner Detection

```python
#sudo apt install libopencv-dev python3-opencv
#pip install opencv-python
#pip install opencv-contrib-python

import cv2 
import matplotlib.pyplot as plt
from data_loader import DataLoader

#load training set:
dataLoader = DataLoader()
dictResult = dataLoader.get_characters_train_data(path = "data/character_set_labeled", num_samples=20,)
image1 = dictResult['He'][13]
plt.imshow(image1)
```

```python
sift = cv2.xfeatures2d.SIFT_create()
keypoints_1, descriptors_1 = sift.detectAndCompute(image1,None)
image1features = cv2.drawKeypoints(image1,keypoints_1,image1)
plt.imshow(image1features)
```

```python
corners = cv2.cornerHarris( image1 ,2,3,0.04 )
corners = cv2.dilate(corners, None, iterations=1)
image1features[corners>0.01*corners.max()] = [255,0,0]
plt.imshow(image1features)
```
